```python
import pandas as pd
```

```python
ds = pd.read_csv("address_nodes.csv")
print("# of address key: " + str(len(ds)))
ds.head(10)
```

```python
ds = pd.read_csv("director_nodes.csv")
print("# of director: " + str(len(ds)))
print("# of unique addr_key: " + str(len(set(ds.addr_key))))
print("# of unique name: " + str(len(set(ds.name))))
ds.head(10)
```

```python
ds = pd.read_csv("corporation_nodes.csv")
print("# of corp_id-address: " + str(len(ds)))
print("# of unique corp_id: " + str(len(set(ds.corp_id))))
print("# of unique addr_key: " + str(len(set(ds.addr_key))))
ds.head(10)
```

```python
ds = pd.read_csv("director_corporation_edges.csv")
print("# of director-corp relationship: " + str(len(ds)))
print("# of unique corp_id: " + str(len(set(ds.corp_id))))
print("# of unique dirr_id: " + str(len(set(ds.dirr_id))))
ds.head(10)
```
